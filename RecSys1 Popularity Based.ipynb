# RecSys1 - Popularity Based
Using the most basic metrics, decide which top 10 book based on their respective ratings.

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read dataset used
books = pd.read_csv (r'C:\Users\user\OneDrive\Desktop\Artificial Intelligence\clean dataset\books.csv',usecols = ['book_id','authors','original_publication_year','title','image_url'])
ratings = pd.read_csv (r'C:\Users\user\OneDrive\Desktop\Artificial Intelligence\clean dataset\ratings.csv')
```

```python
# generating top 10 books based on the average rating of each book 
ratings_count = pd.DataFrame(ratings.groupby(['book_id'])['ratings'].mean())
top10 = ratings_count.sort_values('ratings', ascending = False)

print("Following books are recommended")
sortrating = top10.merge(books, left_index = True, right_on = 'book_id')
sortedrating = sortrating.head(6)
sortedrating
```
